```python
smi = list(truedf[26])
Id = list (truedf[0])
del smi[0]
del Id[0]
m = []
for i in range(len(smi)):
    try:
        m.append(Chem.MolFromSmiles(smi[i]))
    except:
        m.append('')
```

```python
form = []
for i in range(len(m)):
    try:
        form.append(rdMD.CalcMolFormula(m[i]))
    except:
        form.append('')
```

```python
data = {'id':Id, 'MolecularFormula': form }
cn = ['id','MolecularFormula']
formdata = pd.DataFrame(data, columns=cn)
formdata
```
